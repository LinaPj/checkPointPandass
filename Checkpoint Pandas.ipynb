```python
import pandas as pd
import numpy as np
```

```python
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
df=pd.DataFrame(exam_data,labels) #permet de cree des dictionnaires a partir de dictionnaires
df
```

```python
#4 
df.loc["k"]={"name" : "Suresh", "score": 15.5, "attempts": 1, "qualify": "yes"} #AJOUT DE CETTE LIGNE
df
```

```python
#5
df=df.drop("attempts",axis=1) #ECRASE la valeur #pour les lignes axis = 0 pas  1 
df
```

```python
#6
df["success"]=""
for i in df.index:
    if df.loc[i,"score"]>10:
        df.loc[i,"success"]=1
        
    else:
        df.loc[i,"success"]=0
    
df    
```
